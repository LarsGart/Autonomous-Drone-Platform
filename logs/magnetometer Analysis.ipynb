```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'cleanedMagData.txt',names=['C1','C2','C3','C4','C5','C6'],sep=',')
```

```python
df['C1'] = df['C1'].str.replace('(','').astype(float)
df['C3'] = df['C3'].str.replace(')','').astype(float)
df['C4'] = df['C4'].str.replace('(','').astype(float)
df['C6'] = df['C6'].str.replace(')','').astype(float)
```

```python
df.hist(bins=40, figsize=(25,20))
plt.show()
```

```python
import seaborn as sn
from matplotlib.pyplot import figure
figure(figsize=(15, 12))
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
# Import dependencies
import plotly
import plotly.graph_objs as go

# Configure Plotly to be rendered inline in the notebook.
plotly.offline.init_notebook_mode()

# Configure the trace.
trace = go.Scatter3d(
    x=df.C1,  # <-- Put your data instead
    y=df.C2,  # <-- Put your data instead
    z=df.C3,  # <-- Put your data instead
    mode='markers',
    marker={
        'size': 10,
        'opacity': 0.8,
    }
)

# Configure the layout.
layout = go.Layout(
    margin={'l': 0, 'r': 0, 'b': 0, 't': 0}
)

data = [trace]

plot_figure = go.Figure(data=data, layout=layout)

# Render the plot.
plotly.offline.iplot(plot_figure)
```
